```python
import numpy as np

# Create arrays
a = np.array([1, 2, 3, 4, 5])
b = np.array([[1, 2], [3, 4]])

print(a)
print(b)

# Basic operations
print(a.mean())
print(a.std())
print(a * 2)
```

```python
import pandas as pd

data = {
    'Age': [22, 25, 30, 35, 40],
    'Salary': [50000, 60000, 75000, 90000, 110000],
    'Purchased': [0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Filtering
high_salary = df[df['Salary'] > 70000]
print(high_salary)

# Correlation
print(df.corr())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```
